### Figure 4.3: Private equity and hedge fund investment over time by endowment wealth

```stata
ssc install blindschemes
```

```stata
use data/d4_1_nacubo_cae_endow_donations_time, clear
```

```stata
collapse (mean) alternative, by(year pctile)

reshape wide alternative, i(year) j(pctile)

foreach var in alternative {
label var `var'8 "Private >99"
label var `var'7 "Private 95-99"
label var `var'6 "Private 90-95"
label var `var'5 "Private 75-90"
label var `var'4 "Private 0-75"
label var `var'1 "Public 0-95"
label var `var'2 "Public 95-99"
label var `var'3 "Public >99"
}
*
```

```stata
tw (connect alternative* year if year>1999 & year<2014, sort msym(D D D O O O O O) ///
	mc(gs8 gs8 gs8 gs8 gs8 gs10 black black) ///
	mfc(white gs12 gs8 white gs14 gs10 gs6 black) ///
	lp(dot dot dot l l l l l) lc(black black black gs8 gs10 gs10 black black) ///
	lw(medthick medthick medthin thin thin)), ///
	legend(size(small) pos(3)  order(8 7 3 6 2 5 1 4) ///
	subtitle("{bf:Percentile}", size(small))) ///
	ytitle("Share of portfolio in alternative investments") ///
	xsize(6.5) ysize(5) ///
	name(alternative_share, replace) scheme(plotplain)

graph export figures/f4_3_investments_over_time.pdf, replace
```
