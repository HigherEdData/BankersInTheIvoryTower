### Figure 4.1: Mean endowment wealth by endowment wealth percentile

```stata
*ssc install blindschemes
```

```stata
use data/d4_1_nacubo_cae_endow_donations_time, clear
```

```stata
collapse (mean) endow_assets all_under, by(year pctile)

reshape wide endow_assets all_under, i(year) j(pctile)

foreach var in endow_assets all_under {
label var `var'8 "Private >99"
label var `var'7 "Private 95-99"
label var `var'6 "Private 90-95"
label var `var'5 "Private 75-90"
label var `var'4 "Private 0-75"
label var `var'1 "Public 0-95"
label var `var'2 "Public 95-99"
label var `var'3 "Public >99"
}
*

tw (connect endow_assets* year, sort msym(D D D O O O O O) ///
	mc(gs8 gs8 gs8 gs8 gs8 gs10 black black) ///
	mfc(white gs12 gs8 white gs14 gs10 gs6 black) ///
	lp(dot dot dot l l l l l) lc(black black black gs8 gs10 gs10 black black) ///
	lw(medthick medthick medthin thin thin)), ///
	legend(size(small) pos(3) order(8 3 7 2 6 5 4 1) ///
	subtitle("{bf:Endowment wealth}" "{bf:percentile}", size(small))) ///
	ytitle("Endowment assets (2016$ in billions)") ///
	xsize(6.5) ylab(0(5)20) ///
	name(endow_over_time, replace) scheme(plotplain)

graph export figures/f4_1_endowment_values_by_percentile.pdf, replace
```

```stata
tw (connect all_under* year if year>1986, sort msym(D D D O O O O O) ///
	mc(gs8 gs8 gs8 gs8 gs8 gs10 black black) ///
	mfc(white gs12 gs8 white gs14 gs10 gs6 black) ///
	lp(dot dot dot l l l l l) lc(black black black gs8 gs10 gs10 black black) ///
	lw(medthick medthick medthin thin thin)), ///
	legend(size(small) pos(3) order(3 2 1 7 8 6 5 4) ///
 	subtitle("{bf:Endowment wealth}" "{bf:percentile}", size(small))) ///
	ytitle("Mean undergrad enrollment in thousands") ///
	name(giving_student, replace) scheme(plotplain)

graph export figures/f4_a1_undergrad_enrollment_by_endowment_size.pdf, replace
```
